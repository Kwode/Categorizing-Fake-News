```python
import pandas as pd
import spacy
import nltk
```

```python
fake_data = pd.read_csv('fake_news_data.csv')
fake_data.head()
fake_news = fake_data[['text', 'fake_or_factual']]
fake_news.head()
```

POS TAGGING

```python
nlp = spacy.load('en_core_web_sm')
fake_news = fake_data[fake_data['fake_or_factual'] == 'Fake News']
factual_news = fake_data[fake_data['fake_or_factual'] == 'Factual News']

fake_spacy_doc = list(nlp.pipe(fake_news['text']))
factual_spacy_doc = list(nlp.pipe(factual_news['text']))
```

```python
fake_tags = []

for doc in fake_spacy_doc: 
    fake_tags.append([{'token': token.text, 'pos tags': token.pos_, 'ner': token.ent_type_} for token in doc])

fake_tags[1][0]
```
